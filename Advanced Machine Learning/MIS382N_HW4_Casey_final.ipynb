A Bayesian network is a directed acyclic graph (DAG) that captures a subset of the independence relationships of a given joint probability distribution. 

In a Bayesian network $G=(V,E)$, each node $i \in V$in the directed graph corresponds to a random variable and each directed edge $e \in E$ represents a statistical dependence. Each node is associated with a conditional probability distribution of the corresponding random variables given its parents in the DAG. The joint probability distribution factorizes w.r.t the directed graph $G$ if $p(x_1, ..., x_n)=\prod_{i \in V} p(x_i | x_{parents(i)})$, provided the variables are visited in a topologically sorted order. def get_mmd_rbf(X, Y, gamma=1.0):
```
    """MMD using rbf (gaussian) kernel (i.e., k(x,y) = exp(-gamma * ||x-y||^2 / 2))
    Arguments:
        X {[n_sample1, dim]} -- [X matrix]
        Y {[n_sample2, dim]} -- [Y matrix]
    Keyword Arguments:
        gamma {float} -- [kernel parameter] (default: {1.0})
    Returns:
        [scalar] -- [MMD value]
    """
    XX = metrics.pairwise.rbf_kernel(X, X, gamma)
    YY = metrics.pairwise.rbf_kernel(Y, Y, gamma)
    XY = metrics.pairwise.rbf_kernel(X, Y, gamma)
    return XX.mean() + YY.mean() - 2 * XY.mean()
```

````python
# adult census dataset, for Section A and Section B
def load_adult():
  df = pd.read_csv('adult.csv')
  df = pd.get_dummies(df)
  X = df.drop(['target'], axis=1)
  y = df[['target']]
  return X, np.reshape(y, (-1, ))

X, y = load_adult()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
````

````python

from sklearn.svm import LinearSVC

#part 1 
class NaivelyCalibratedLinearSVC(LinearSVC):
    """LinearSVC with `predict_proba` method that naively scales
    `decision_function` output."""

    def fit(self, X, y):
        super().fit(X, y)
        df = self.decision_function(X)
        self.df_min_ = df.min()
        self.df_max_ = df.max()

    def predict_proba(self, X):
        """Min-max scale output of `decision_function` to [0,1]."""
        df = self.decision_function(X)
        calibrated_df = (df - self.df_min_) / (self.df_max_ - self.df_min_)
        proba_pos_class = np.clip(calibrated_df, 0, 1)
        proba_neg_class = 1 - proba_pos_class
        proba = np.c_[proba_neg_class, proba_pos_class]
        return proba

#part 2
from sklearn.calibration import calibration_curve, CalibrationDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression


# Create classifiers
lr = LogisticRegression()
mlp = MLPClassifier()
#svc = NaivelyCalibratedLinearSVC(C=1.0)
rfc = RandomForestClassifier()

clf_list = [
    (lr, "Logistic"),
    (mlp, "MLP"),
    #(svc, "SVC"),
    (rfc, "Random Forest"),
]

#part 3

import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(10, 10))
gs = GridSpec(4, 2)
colors = plt.cm.get_cmap("Dark2")

ax_calibration_curve = fig.add_subplot(gs[:2, :2])
calibration_displays = {}
for i, (clf, name) in enumerate(clf_list):
    clf.fit(X_train, y_train)
    display = CalibrationDisplay.from_estimator(
        clf,
        X_test,
        y_test,
        n_bins=10,
        name=name,
        ax=ax_calibration_curve,
        color=colors(i),
    )
    calibration_displays[name] = display

ax_calibration_curve.grid()
ax_calibration_curve.set_title("Calibration plots - Adult Census")

# Add histogram
grid_positions = [(2, 0), (2, 1), (3, 0), (3, 1)]
for i, (_, name) in enumerate(clf_list):
    row, col = grid_positions[i]
    ax = fig.add_subplot(gs[row, col])

    ax.hist(
        calibration_displays[name].y_prob,
        range=(0, 1),
        bins=10,
        label=name,
        color=colors(i),
    )
    ax.set(title=name, xlabel="Mean predicted probability", ylabel="Count")

plt.tight_layout()
plt.show()
````

````python
len_cont = len(desired_columns_to_convert)
def dist_c(query_instance,d,cfs,cont_len):
        
    mad = pd.Series(d.get_mads())
    mad[mad == 0] = 1 # to prevent infinite
   
    dist_c = 0
    for idx,row in cfs.iterrows():
        dist_c += np.abs(row-query_instance)/mad
       
    return dist_c.mean()/cont_len

def get_mmd_rbf(cfs_dict, gamma=1.0):
    """MMD using rbf (gaussian) kernel (i.e., k(x,y) = exp(-gamma * ||x-y||^2 / 2))
    Arguments:
        X {[n_sample1, dim]} -- [X matrix]
        Y {[n_sample2, dim]} -- [Y matrix]
    Keyword Arguments:
        gamma {float} -- [kernel parameter] (default: {1.0})
    Returns:
        [scalar] -- [MMD value]
        
    """
    XX_list = []
    XY_list = []
    YY_list = []
    
    for i in range(len(cfs_dict)):
        X = cfs_dict[i][0].values.reshape(-1, 1).T
        Y = cfs_dict[i][1].drop(columns = "target").values
        XX = metrics.pairwise.rbf_kernel(X, X, gamma)
        YY = metrics.pairwise.rbf_kernel(Y, Y, gamma)
        XY = metrics.pairwise.rbf_kernel(X, Y, gamma)
        XX_list.append(XX)
        XY_list.append(XY)
        YY_list.append(YY)

    return  np.mean(XX_list) + np.mean(YY_list) - 2 * np.mean(XY_list)

def cont_prox(cfs_dict,d,len_cont):
    cont_prox = 0
    for i in range(len(cfs_dict)):
        cont_prox+= dist_c(cfs_dict[i][0],d,cfs_dict[i][1],len_cont)
       
    return - cont_prox/len(cfs_dict)

def diversity(cfs_dict,d,len_cont):
    divers = 0
    for i in range(len(cfs_dict)):
        cfs = cfs_dict[i][1]
        for j in range(len(cfs)):
            row = cfs_dict[i][1].iloc[j]
            
            divers+= dist_c(row,d,cfs,len_cont)/(cfs*cfs).sum().mean()
    return divers

def validity(cfs_dict):
    r_outer = []
    for i in range(len(cfs_dict)):
        unique_cfs = len(set(map(str,cfs_dict[i][1].values)))
        r_outer.append(float(unique_cfs) / len(cfs_dict[i][1]))
    return np.mean(r_outer)

    
def sparsity(cfs_dict,d): 
    
    dist_outer = []

    for i in range(len(cfs_dict)):
        p1 = cfs_dict[i][0]
        cfs = cfs_dict[i][1]
        dist_outer.append(1 - dist_c(p1,d,cfs,len(p1.T)))
    return np.mean(dist_outer)
````

````python
#creating rfc counterfactual dictionaries 
cfs_random = {}
cfs_genetic = {}               
for  i in range(len(query_instances)):
      cf_random = random_rfc.cf_examples_list[i].final_cfs_df
      cf_genetic = genetic_rfc.cf_examples_list[i].final_cfs_df

            
      cfs_random[i] = (query_instances.iloc[i],cf_random)
      cfs_genetic[i] = (query_instances.iloc[i],cf_genetic)


#Random Forest - DiCE Random Metrics 
random_mmd = get_mmd_rbf(cfs_random, gamma=1.0)
random_proximity = cont_prox(cfs_random,d,len_cont)
random_diversity = diversity(cfs_random,d,len_cont)
random_validity = validity(cfs_random)
random_sparsity = sparsity(cfs_random,d)

#Random Forest - DiCE Genetic Metrics 
#genetic_mmd = get_mmd_rbf(cfs_genetic, gamma=1.0)
#genetic_proximity = cont_prox(cfs_genetic,d,len_cont)
#genetic_diversity = diversity(cfs_genetic,d,len_cont)
#genetic_validity = validity(cfs_genetic)
#genetic_sparsity = sparsity(cfs_genetic,d)

#creating metrics dataframe
metrics_df = pd.DataFrame(columns=['Model', 'MMD', 'Proximity', 'Diversity', 'Validity', 'Sparsity'])

#adding metrics to dataframe
metrics_df.loc[0] = ['Adult - RFC - DiCE Random',random_mmd, random_proximity,random_diversity,random_validity,random_sparsity ]
metrics_df.loc[1] = ['Adult - RFC - DiCE Genetic',genetic_mmd, genetic_proximity,genetic_diversity,genetic_validity,genetic_sparsity ]
````
